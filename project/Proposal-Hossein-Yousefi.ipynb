# HWRS640 - Final Project: Proposal Pitch

## Learning to Fuse 32-Member Stochastic Downscaling Ensembles for Improved Extreme Rainfall Prediction

Hossein Yousefi Sohi

## 1. Motivation and Scientific Question

High-resolution precipitation downscaling is especially difficult for convective and monsoon events, where localized extremes are easily smoothed out. In my current workflow, a stochastic downscaling model produces 32 possible high-resolution precipitation fields for each time step, but operational evaluation usually collapses these members with simple rules (ensemble mean, median, or quantile).

**Scientific question:** Can a second-stage neural network learn to combine all 32 ensemble members into a better precipitation forecast (and better calibrated uncertainty) than standard ensemble averaging, especially for heavy-rain thresholds?

## 2. Dataset

I will use real WRF-based gridded weather data already available in my research environment, including:

- **Inputs:** coarse-resolution WRF fields (12 km) regridded to a 3 km analysis grid
- **Targets:** high-resolution WRF fields (3 km) on the same grid
- **Model outputs:** 32-member stochastic downscaling ensembles at 3 km resolution
- **Primary variable:** total precipitation (RAIN_TOTAL)
- **Domain/time:** western U.S. with Arizona-focused diagnostics; 2000–2012 archive with 2012 held for validation/testing

The data are stored as large spatiotemporal Zarr/NetCDF fields. Known challenges include:

- Large file I/O and memory limits
- Strong class imbalance for heavy precipitation
- Land/ocean masking and quality-control filtering
- Careful time-based split to avoid leakage

## 3. Proposed Method

The central course method is **deep learning** (CNN-based post-processing / stacking model).

I will build a meta-ensemble neural network that learns from all 32 ensemble members simultaneously rather than using fixed reducers.

**Planned setup:**

- **Input tensor:** 32 precipitation members per time step (plus optional summary features: member mean/std/quantiles, and optional coarse input field)
- **Architecture (preliminary):** a lightweight CNN with permutation-invariant ensemble fusion (DeepSets-style aggregation or shared member encoder + feature pooling)
- **Outputs:**
  - deterministic best-estimate precipitation map
  - exceedance probabilities for heavy-rain thresholds (e.g., >1, >5, >10 mm)
- **Losses:** Huber/MSE for deterministic output + binary cross-entropy (or focal loss) for threshold exceedance probability heads

This is a new analysis angle for my workflow: instead of improving the base stochastic model itself, I am testing whether learned ensemble fusion can recover additional skill from existing ensemble information.

## 4. Expected Outcomes and Evaluation

I expect the learned fusion model to outperform simple ensemble reducers for extreme precipitation detection while keeping comparable bulk error on non-extreme cases.

I will evaluate against baselines (ensemble mean, median, q90, and best-single-member reference) using:

- **Deterministic metrics:** RMSE, MAE, PBIAS, Pearson correlation
- **Extreme-event metrics:** CSI at 1/5/10 mm and FSS across spatial scales
- **Probabilistic metrics:** CRPS/Brier score and reliability curves
- **Diagnostics:** spread-skill relationship and event-based map comparisons for top rainy cases

A successful project will show clear, interpretable gains in at least one extreme-rainfall metric (e.g., CSI/FSS) without severe degradation in overall bias/error, and will identify where learned fusion helps most (e.g., monsoon events and localized heavy-rain structures).
